<H3>MINI BATCH GD FROM SCRATCH

```python
from sklearn.datasets import load_diabetes
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X,y=load_diabetes(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred=reg.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
class MBGDRegressor:
    def __init__(self,batchsize,learning_rate=0.1,epoch=50):
        self.lr=learning_rate
        self.epoch=epoch
        self.intercept_=None
        self.coef_=None
        self.batchsize=batchsize
    def fit(self,X_train,Y_train):
        self.intercept_=0
        self.coef_=np.ones(X_train.shape[1])
        for i in range(self.epoch):
            for j in range(int(X_train.shape[0]/self.batchsize)):
                idx=random.sample(range(X_train.shape[0]),self.batchsize)
                y_hat=np.dot(X_train[idx],self.coef_)+self.intercept_
                slope_intercept=-2*np.mean(y_train[idx]-y_hat)
                self.intercept_=self.intercept_-(self.lr*slope_intercept)
                slope_coef=-2*np.dot((y_train[idx]-y_hat),X_train[idx])
                self.coef_=self.coef_-(self.lr*slope_coef)
        print(self.intercept_,self.coef_)
    def predict(self,X_test):
        return np.dot(X_test,self.coef_)+self.intercept_
        
#here the code is combination of both the batch gradient descent and the  stochastic gd
    
```

```python
import random
random.sample(range(1,100),10)
```

```python
mbr=MBGDRegressor(batchsize=int(X_train.shape[0]/10))
```

```python
mbr.fit(X_train,y_train)
```

```python
y_pred=mbr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
