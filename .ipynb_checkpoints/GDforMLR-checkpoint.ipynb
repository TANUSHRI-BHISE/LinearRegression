```python
from sklearn.datasets import load_diabetes
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X,y=load_diabetes(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred=reg.predict(X_test)
```

```python

from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

<h3>IMPLEMENTATION OF MULTIPLE LINEAR REGRESSION USING GD FROM SCRATCH
    

```python
class GDRegressor:
    def __init__(self,learning_rate=0.1,epoch=100):
        self.coef_=None
        self.intercept_=None
        self.lr=learning_rate
        self.epoch=epoch
    def fit(self,X_train,y_train):
        self.intercept_=0
        self.coef_=np.ones(X_train.shape[1])
        for i in range(self.epoch):
            #update all the intercepts and the coefficient
            #intercept=intercent-n(slope)
            #slope=-2/n sumation (y-yhat)
            y_hat=np.dot(X_train,self.coef_)+self.intercept_
            slope_intercept=-2*np.mean(y_train-y_hat)
            self.intercept_=self.intercept_-(self.lr*slope_intercept)
            
            slope_coef=-2*np.dot((y_train-y_hat),X_train)/X_train.shape[0]
            self.coef_=self.coef_-(self.lr*slope_coef)
        print(self.intercept_,self.coef_)
        #print(y_hat.shape)

    def predict(self,X_test):
        return np.dot(X_test,self.coef_)+self.intercept_
    
```

```python
gd=GDRegressor(0.1,100)
```

```python
gd.fit(X_train,y_train)
```
