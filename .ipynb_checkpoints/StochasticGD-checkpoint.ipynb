<h3>SCRATCH IMPLEMENTATION OF STOCHASTIC GRADIENT DESCENT</h3>

```python
from sklearn.datasets import load_diabetes
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X,y=load_diabetes(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred=reg.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
class SGDRegressor:
    def __init__(self,learning_rate=0.1,epoch=100):
        self.coef_=None
        self.intercept_=None
        self.lr=learning_rate
        self.epoch=epoch
    def fit(self,X_train,y_train):
        self.intercept_=0
        self.coef_=np.ones(X_train.shape[1])
        for i in range(self.epoch):
            for j in range(X_train.shape[0]):
                #genrate a random number
                idx=np.random.randint(0,X_train.shape[0])
                y_hat=np.dot(X_train[idx],self.coef_)+self.intercept_
                slope_intercept=-2*(y_train[idx]-y_hat)
                self.intercept_=self.intercept_-(self.lr*slope_intercept)
                slope_coef=-2*np.dot((y_train[idx]-y_hat),X_train[idx])
                self.coef_=self.coef_-(self.lr*slope_coef)
                
        print(self.intercept_,self.coef_)
        #print(y_hat.shape)

    def predict(self,X_test):
        return np.dot(X_test,self.coef_)+self.intercept_
    
```

```python
sgd=SGDRegressor(learning_rate=0.01,epoch=50)
sgd.fit(X_train,y_train)
```

```python
y_pred=sgd.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
# ‚úÖ Batch Gradient Descent

# 1 epoch ‚Üí only 1 update

# Means:

# you update M and C (or m and b) only one time

# because gradient is calculated using all rows together here in error function we use summation

# ‚úÖ So:

# Updates per epoch
# =
# 1
# Updates per epoch=1
# ‚úÖ Stochastic Gradient Descent (SGD)

# 1 epoch ‚Üí updates happen many times that is for 1 epoch we run inner j loop to update m and c 
#     for randomly choosen values for number of row times

# Specifically:

# you update M and C for number of rows times in 1 epoch ,here error fuction do not uses summation

# because each update uses only 1 row

# ‚úÖ So if dataset has 
# ùëÅ
# N rows:

# Updates per epoch
# =
# ùëÅ
# Updates per epoch=N
```
