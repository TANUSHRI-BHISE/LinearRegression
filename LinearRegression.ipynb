```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("placement.csv")
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('Cgpa')
plt.ylabel('package in lpa')
```

```python
X=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
lr=LinearRegression()
lr
```

```python
lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
#sklearn provides the mse,mae r2 score matrix all 
y_pred=lr.predict(X_test)
```

```python
y_test.values
```

```python
print("mae",mean_absolute_error(y_test,y_pred))#in actual formula we do actual value-predicted value
#unit is same as y that is lpa
```

```python
print("mse",mean_squared_error(y_test,y_pred))#not in lpa
```

```python
print("rmse",np.sqrt(mean_squared_error(y_test,y_pred)))#in lpa
```

```python
print("r2score",r2_score(y_test,y_pred))
r2=r2_score(y_test,y_pred)
```
