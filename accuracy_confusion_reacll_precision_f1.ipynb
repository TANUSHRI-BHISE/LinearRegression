```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("heart_disease.csv")
```

```python
print("Original shape:", df.shape)
print(df.head())
```

```python
print("\nMissing values:\n", df.isnull().sum())
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
print("\nNumeric columns:", num_cols)
print("\nCategorical columns:", cat_cols)
```

```python
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
```

```python
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

```python
print("\nData types after conversion:\n", df.dtypes)
print("\nMissing after cleaning:\n", df.isnull().sum())
```

```python

df.to_csv("heart_clean_numeric.csv", index=False)

print("\n✅ Saved as heart_clean_numeric.csv")
print(df.head())
```

```python
X=df.iloc[:,0:-1]
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
```

```python
clf1.fit(X_train,y_train)
```

```python
clf2.fit(X_train,y_train)
```

```python
y_pred1=clf1.predict(X_test)
y_pred2=clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("accuracy of logistic",accuracy_score(y_test,y_pred1))
print("accuracy of dt",accuracy_score(y_test,y_pred1))
```
